```python
#Query to get all Covid Tweets about American response

c1 = twint.Config()

c1.Search = '#Covid-19 and America and Pandemic'

c1.Since = '2020-01-01'
c1.Until = '2020-07-28'
c1.Lang = 'en'

c1.Store_json = True
c1.Output = 'Covid_tweets3.json'
```

```python
twint.run.Search(c1)
```
