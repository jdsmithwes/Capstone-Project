```python
#Libraries to get tweets
import twint
import nest_asyncio

nest_asyncio.apply()
```

```python
#Query to get all Covid Tweets about American response

c = twint.Config()

c.Search = '#Covid-19 or Covid-19 and America'

c.Since = '2020-01-01'
c.Until = '2020-07-20'
c.Lang = 'en'

c.Store_json = True
c.Output = 'Covid_tweets3.json'
```

```python
twint.run.Search(c)
```
