```python
# Query to get Tweets from NY Governor Andrew Cuomo
c = twint.Config()

c.Username = 'nygovcuomo'
c.Search = 'Covid-19'

c.Since = '2020-01-01'
c.Until = '2020-07-10'
c.Lang = 'en'

c.Store_json = True
c.Output = 'Cuomo_Covid_tweets3.json'
```

```python
twint.run.Search(c)
```
